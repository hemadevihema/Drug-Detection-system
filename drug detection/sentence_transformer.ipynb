```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from pprint import pprint
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
Sentences = ['Laptops are evolving at a dramatic pace, and this is the place to track their progress, or lack thereof. The best laptop is out there, and our laptop reviews dig deep into what’s new from the world’s biggest manufacturers to help you find it. From each new iteration of the Apple MacBook and Microsoft Surface to what’s coming up from likes of Dell, HP, Lenovo, and more, The Verge has you covered.']
```

```python
Sentence_embeddings= model.encode(Sentences)
```

```python
for Sentences, embedding in zip(Sentences, Sentence_embeddings):
  print("Sentence:" , Sentences)
  print("Embedding:"  , embedding)
  print("")
```

```python
len(Sentence_embeddings[0])
```

```python
pprint('Similarity between {} and {} is {}'.format(Sentences[0],
                                                   Sentences[1],
                                                   cosine_similarity(Sentence_embeddings[0].reshape(1,-1),
                                                                     Sentence_embeddings[0].reshape(1,-1))[0][0]))
```
